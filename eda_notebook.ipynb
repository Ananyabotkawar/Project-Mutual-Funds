```python
import os
import pandas as pd

# Updated to match your exact file name
master_path = os.path.join("data", "raw", "01_fund_master.csv")

if not os.path.exists(master_path):
    print(f"Error: Could not find the file at {master_path}")
    print("Please check if the file is inside the data/raw folder.")
else:
    # Read the CSV
    df_master = pd.read_csv(master_path)
    
    # Clean up column names by removing any accidental spaces
    df_master.columns = df_master.columns.str.strip()
    
   
    print("FUND MASTER EXPLORATORY DATA ANALYSIS")
  
    
    # 1. Unique Fund Houses
    if 'fund_house' in df_master.columns:
        print(" UNIQUE FUND HOUSES ")
        print(df_master['fund_house'].unique())
    else:
        print("⚠️ 'fund_house' column not found.")
        
    print("\n" + "-"*40 + "\n")
    
    # 2. Unique Categories
    if 'category' in df_master.columns:
        print("UNIQUE CATEGORIES ")
        print(df_master['category'].unique())
    else:
        print("⚠️ 'category' column not found.")
        
    print("\n" + "-"*40 + "\n")
    
    # 3. Unique Sub-Categories
    if 'sub_category' in df_master.columns:
        print(" UNIQUE SUB-CATEGORIES ")
        print(df_master['sub_category'].unique())
    else:
        print("⚠️ 'sub_category' column not found.")
        
    print("\n" + "-"*40 + "\n")
    
    # 4. Unique Risk Grades
    if 'risk_grade' in df_master.columns:
        print("--- UNIQUE RISK GRADES ---")
        print(df_master['risk_grade'].unique())
    else:
        # If your column has a slightly different name like 'risk' or 'risk_rating'
        possible_risk_cols = [c for c in df_master.columns if 'risk' in c.lower()]
        if possible_risk_cols:
            print(f"--- UNIQUE VALUES FOR {possible_risk_cols[0].upper()} ---")
            print(df_master[possible_risk_cols[0]].unique())
        else:
            print("⚠️ Risk column not found. Available columns are:")
            print(df_master.columns.tolist())

print("==================================================")
```
